```python
import math
import pandas as pd
```

```python
def corr_lin_test_t(x , y):
  moy_x = sum(x)/len(y)
  moy_y = sum(y)/len(y)
  cov_xy = sum((a - moy_x)*(b - moy_y) for (a,b) in zip(x , y))/ len(x)
  etX = (sum((a - moy_x)**2 for a in x)/len(x))**0.5
  etY = (sum((b - moy_y)**2 for b in y)/len(y))**0.5
  r = round(cov_xy/(etX*etY),3)
  sr = math.sqrt((1-r**2)/(len(x)-2))
  t = round((abs(r))/sr,3)
  ddl = len(x)-2
  return('Corrélation linéaire est : ' ,r , 'La valeur de t est : ' ,t , 'DDL = ' ,ddl ,'alpha = 0.05')

  
```

Calculer Correlation Lineaire entre les VAR

```python
corr_lin_test_t(x_attitude , y_duree)
```

```python
corr_lin_test_t(x_attitude , z_importance)
```

```python
corr_lin_test_t(z_importance , y_duree)
```

```python
df = pd.read_excel('/content/add1.xlsx')
df.head(13)
```

```python
import numpy as np
matrix = df.corr()
matrix
```

```python
sr_value = []
t_value = []

for i in matrix.values.reshape(-1):
  if i == 1 :
    pass
  else:
    sr_value.append(math.sqrt((1-i**2)/10))
    t_value.append(((abs(i))/math.sqrt((1-i**2)/10)))
```

```python
len(sr_value)
```

```python
data ={'Sr' : sr_value,'T' : t_value}
df_st = pd.DataFrame(data)
```

```python
var_c =[]
for i in matrix.index:
  for j in matrix.index:
    if i == j:
      pass
    else:
      var_c.append((i,j))
```

```python
len(var_c)
```

```python
df_st['variables_de_liaison'] = var_c
```

```python
t_seuil = 1.8125
interp = []
for i in df_st['T']:
  if abs(i) >= t_seuil:
    interp.append('Rejet H0 , liaison linéaire significatif')
  else:
    interp.append('Non rejet de H0 , pas de liaison linéaire')
```

```python
len(interp)
```

```python
df_st['Interpretation'] = interp
```
